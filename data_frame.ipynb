```python
import os
import pandas as pd
```

```python
path=os.getcwd()
data=pd.read_csv(os.path.join(path,'data','wave_era5_2015.csv'))
print(data)
```

## Exploring data frame


```python
print(data.columns)
```

```python
time=data['Time']
print(time)
```
