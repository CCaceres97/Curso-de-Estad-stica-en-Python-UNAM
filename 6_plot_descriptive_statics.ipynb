```python
import os
def import_csv(file):
    # Read the file into a DataFrame: df
    import pandas as pd
    return pd.read_csv(file)
```

```python
# Path to the file to be imported
path = os.getcwd()

# Import the file
CO2Data = import_csv(os.path.join(path, 'Terminos_lagoon_TA_DIC_2023_RawData.csv'))
```

```python
#PLot simple histogram
import matplotlib.pyplot as plt

# Select data
sal_data = CO2Data["sal_psu"]

plt.hist(sal_data, bins=10)
plt.xlabel("Salinity (PSU)")
plt.ylabel("Probablity")
plt.show()
```

```python
# Plot the histogram with density (nomalized)
# Select data 
data = CO2Data["sal_psu"]


plt.hist(data, bins=10, density=True)
plt.xlabel("Salinity (PSU)")
plt.ylabel("Probablity")
plt.show()
```

```python
# Creating plot

do_data = CO2Data["do_mg_l"]
plt.boxplot(do_data)

plt.xlabel("DO (mg/L)")
plt.ylabel("Concentration")
plt.title("Boxplot of DO Concentration")

# show plot
plt.show()
```

```python
# Plot boxplot for salinity data
data = CO2Data["sal_psu"]

fig2, ax2 = plt.subplots()

ax2.set_title('Notched boxes')
ax2.boxplot(data, notch=True)

plt.ylabel('Salinity ($PSU$)', fontsize = 12)
plt.show()
```

```python
# Plot boxplot for TA data
green_diamond = dict(markerfacecolor='g', marker='D')
fig3, ax3 = plt.subplots()

ax3.set_title('Changed Outlier Symbols')

ax3.boxplot(data, flierprops=green_diamond)


plt.ylabel('TA ($\mu mol  \; kg^{-1}$)', fontsize = 12)
plt.show()
```

## Tarea

```python
#plot histogram of DIC and TA
import seaborn as sns
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
sns.histplot(CO2Data['dic_micromol_kg'], bins=20, kde=True, ax=ax[0], color='blue')
ax[0].set_title('Histogram of DIC')
ax[0].set_xlabel('DIC ($\mu mol \; kg^{-1}$)')
ax[0].set_ylabel('Frequency')
sns.histplot(CO2Data['ta_micromol_kg'], bins=20, kde=True, ax=ax[1], color='orange')
ax[1].set_title('Histogram of TA')
ax[1].set_xlabel('TA ($\mu mol \; kg^{-1}$)')
ax[1].set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```
